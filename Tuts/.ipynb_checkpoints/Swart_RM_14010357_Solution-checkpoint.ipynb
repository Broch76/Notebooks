```python
from scipy.optimize import fsolve
from matplotlib import pyplot as plt
import numpy as np
from scipy.integrate import odeint
%matplotlib inline
```

```python
"""a)"""
def Kval(D):
    FT = FAo = 15 #mol/mi
    A = (np.pi/4)*D**2
    G = (FT/60)*50/1000/A
    E = 0.2
    Po = 10*(1000*100)
    dp = 0.005
    mu = 9e-5
    FT = FAo = 15 #mol/mi
    k1, k2, k3 = 0.01, 0.05, 0.0005
    rhoA = (Po/1000)*30/(8.314*(273.15+300))
    rhoB = 1000
    K = (1/(rhoB*A))*(-G/(rhoA*dp))*((1-E)/(E**3))*(150*mu*(1-E)/dp + 1.75*G)
    return K

K = Kval(0.1)

def PBR(F,t):
    FA, FB, FC, FD, P = F
    Ftot = (FA + FB + FC + FD)
    Pr = P/(100000)
    ra1 = -k1*(FA/Ftot)*Pr
    ra2 = -k2*(FA/Ftot)*Pr*(FB/Ftot)*Pr
    ra3 = -k3*(FA/Ftot)*Pr*(FC/Ftot)*Pr
    dFadw = ra1 + ra2 + ra3
    dFbdw = -ra1 + ra2
    dFcdw = -ra2 + ra3
    dFddw = -ra3
    dPdw = K*(Po*Ftot)/(P*FT)
    return (dFadw, dFbdw, dFcdw, dFddw,dPdw)

wspan = np.linspace(0,150,1000)
F = odeint(PBR,[FAo,0,0,0,Po],wspan)
Fa = F[:,0]
Fb = F[:,1]
Fc = F[:,2]
Fd = F[:,3]
P = F[:,4]

X = (FAo- Fa)/FAo
W = np.interp(5.5,Fc,wspan)
Pf = np.interp(5.5,Fc,P)

print ('W = ' +str(W))
print ('P = ' +str(Pf/(100000)))
plt.plot(wspan,Fa, label = "A")
plt.plot(wspan,Fb, label = "B")
plt.plot(wspan,Fc, label = "C")
plt.plot(wspan,Fd, label = "D")
plt.legend(loc = "best")
plt.show()
plt.plot(wspan,X)
plt.show()
plt.plot(wspan,P/(100000))
plt.show()
```

```python
"""b)"""
def Kval(D):
    FT = FAo = 15 #mol/mi
    A = (np.pi/4)*D**2
    G = (FT/60)*50/1000/A
    E = 0.2
    Po = 10*(1000*100)
    dp = 0.005
    mu = 9e-5
    FT = FAo = 15 #mol/mi
    k1, k2, k3 = 0.01, 0.05, 0.0005
    rhoA = (Po/1000)*30/(8.314*(273.15+300))
    rhoB = 1000
    K = (1/(rhoB*A))*(-G/(rhoA*dp))*((1-E)/(E**3))*(150*mu*(1-E)/dp + 1.75*G)
    return K

def func(D):
    K = Kval(D)
    def PBR(F,t):
        FA, FB, FC, FD, P = F
        Ftot = (FA + FB + FC + FD)
        Pr = P/(100000)
        ra1 = -k1*(FA/Ftot)*Pr
        ra2 = -k2*(FA/Ftot)*Pr*(FB/Ftot)*Pr
        ra3 = -k3*(FA/Ftot)*Pr*(FC/Ftot)*Pr
        dFadw = ra1 + ra2 + ra3
        dFbdw = -ra1 + ra2
        dFcdw = -ra2 + ra3
        dFddw = -ra3
        dPdw = K*(Po*Ftot)/(P*FT)
        return (dFadw, dFbdw, dFcdw, dFddw,dPdw)
    
    wspan = np.linspace(0,100,1000)
    F = odeint(PBR,[FAo,0,0,0,Po],wspan)
    P = F[:,4]
    return (P[-1]/(100000) - 6)

D = fsolve(func,[0.1])
print ("Diameter (P @ 6) = " + str(D[0]) + "m")
```
